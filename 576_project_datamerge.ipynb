```
df_f = df_f[df_f['cancelled'] == 0]
df_f = df_f[df_f['origin'] == 'ORD']
```

Only filtering out flights departing from ORD. Since cancelled flights have no overlap with weather delay. We remove observations belonging to cancelled flight.

```
df_w = df_w_origin.drop(['year', 'month', 'day', 'hour'], axis=1)
```

Only filtering out weather in ORD airport

```
df_f['departure_time'] = pd.to_datetime(df_f[[ 'year', 'month', 'day', 'hour','minute']])
```

In the original dataset, we have multiple columns -- 'year', 'month', 'day', 'hour', 'minute' recording the detailed departure time of flight, we also have another column 'dep_time' presenting the depature hour and minute. This make dataset redundent and hard to read. In this step, I'm trying to use single column to display detailed and departure time.

```
df_f = df_f.reset_index().drop(['index'], axis=1)
```

```
def delay_minute(x):
    return datetime.timedelta(minutes=x)
```

```
df_f['delay'] = pd.DataFrame(map(delay_minute,df_f['dep_delay']))
```

Convert 'dep_delay' to computer understandable minutes formate in new column 'delay'

```
df_f['scheduled_dep'] = df_f['departure_time'] - df_f['delay']
```

Scheduled departure time is calculated by **departure_time - delay** , store it in new column 'scheduled_dep'

```
df_f = df_f.drop(df_f.index[df_f[df_f.scheduled_dep <= '2014-01-01 00:00:00'].index])
```

Now we got scheduled departure time, but three of data points falls into 2013 where we don't have corresponding weather data, so I removed these three observations from data frame

```
df_w['wea_time'] = pd.to_datetime(df_w['time_hour'])
```

Also generate a new column 'wea_time' in weather data frame by using 'time_hour' column and using .to_datetime function to convert it to datetime format, which display the hour of weather recorded

## Merge flight and weather data using 'departure_time' as a key

```
df_f = df_f.set_index('scheduled_dep')
```

```
df_w = df_w.set_index('wea_time')
```

```
df = df_f.join(df_w.reindex(df_f.index, method='nearest'))
```

Our merging logic is use flight 'scheduled-dep' to find nearest 'wea_time', and then the corresponding weather data with be pasted. 
For example, 
*   2014-01-01 05:09:00 scheduled departure time align with 2014-01-01 05:00:00 wea_time
*   2014-01-01 06:30:00 scheduled departure time align with 2014-01-01 07:00:00 wea_time
*   2014-01-01 21:55:00 scheduled departure time align with 2014-01-01 22:00:00 wea_time
